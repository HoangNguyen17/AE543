## Problem 2 c

```python
#================ Parameter =================
m,k,X = 1,1,2.5
kc_hardening = 0.5
kc_softening = -0.5
#=============== Computing ==================
X =np.linspace(0,X,100)
w_n = [np.sqrt(k/m)]*len(X) 
w_res_hard = w_n * np.sqrt(np.maximum(0.0, 1 + ((3*kc_hardening*X**2)/(4*k))) )
w_res_soft = w_n * np.sqrt(np.maximum(0.0, 1 + ((3*kc_softening*X**2)/(4*k))) )
```

```python
plt.figure(figsize = (7,5))
plt.grid(True)
plt.plot(X,w_res_soft, color ='red')
plt.plot(X,w_res_hard, color ='blue')
plt.plot(X,w_n, linestyle = '--')


plt.xlabel('Amplitude $X$')
plt.ylabel(r'Backbone frequency $\omega_{res}(X)$')
plt.title('Duffing Backbone Response')

plt.legend([r'Dynamic softening $k_c<0$', r'Dynamic hardening $k_c>1$',r'Linear resonance $\omega_n$'])
plt.savefig('2c.png', dpi=300, bbox_inches='tight')
plt.show()
```
