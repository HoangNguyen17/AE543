## Approach 1.0: Simplify the wing cross section

### Part a: 
Given the mode shape $\phi(x)$ function as following:
$$\phi(x) = 0.5 \left(\frac{x}{L}\right)^2 \left(3 - \frac{x}{L}\right) = \frac{3}{2}\frac{x^2}{L^2} - \frac{1}{2}\frac{x^3}{L^3}$$
$$\phi'(x) = \frac{3x}{L^2} - \frac{3x^2}{2L^3}$$
$$\phi''(x) = \frac{3}{L^2} - \frac{3x}{L^3} = \frac{3}{L^2}\left(1 - \frac{x}{L}\right)$$

At $x= L$, we have 
$$\phi(L) = 0.5(1)^2(3-1) = 1$$
$$\phi'(L) = \frac{3}{L} - \frac{1.5}{L} = \frac{1.5}{L}$$
$$\phi''(x) = \frac{3}{L^2} - \frac{3x}{L^3} = \frac{3}{L^2}\left(1 - \frac{x}{L}\right)$$

* Kinetic energy $\mathcal{T} = \mathcal{T}_{beam} + \mathcal{T}_{tip}$

\begin{equation*}
    \begin{aligned}
        \mathcal{T} &= \mathcal{T}_{beam} + \mathcal{T}_{tip} \\
        \mathcal{T} &= \frac{1}{2} m_w \int_0^L [\phi(x)]^2 dx + \frac{1}{2} M_1 [\phi(L)]^2 + \frac{1}{2} J_{tip} [\phi'(L)]^2\\
        \mathcal{T} &= \frac{1}{2} m_w\int_0^L \left[ \frac{1}{2} \left(\frac{x}{L}\right)^2 \left(3 - \frac{x}{L}\right) \right]^2 dx  +  \frac{1}{2} M_1 (1)^2 +J_{tip} \left(\frac{1.5}{L}\right)^2 \\
        \mathcal{T} &=  m_w \frac{33L}{140} + M_1(1)^2 + J_{tip} \left(\frac{1.5}{L}\right)^2 \\
        \mathcal{T} &= \frac{1}{2} \left[] m_w \frac{33L}{140} + M_1(1)^2 + J_{tip} \left(\frac{1.5}{L}\right)^2 \right]
    \end{aligned}
\end{equation*}

* Potential energy $\mathcal{V} = \mathcal{V}_{beam} + \mathcal{V}_{Springs}$
\begin{equation*}
    \begin{aligned}
        \mathcal{V} &=  \frac{1}{2} EI \int_0^L [\phi''(x)]^2 dx + \frac{1}{2} K_1 [\phi(L)]^2 + \frac{1}{2} K_t [\phi'(L)]^2 \\
        \mathcal{V} &=  \frac{1}{2} EI \int_0^L \left[ \frac{9}{L^4}\left(1 - \frac{x}{L}\right)^2 \right] dx + \frac{1}{2} K_1 (1)^2 + \frac{1}{2} K_t \left(\frac{1.5}{L}\right)^2\\
        \mathcal{V} &= \frac{9EI}{2L^4} \left[ \frac{L}{3} \right]  +\frac{1}{2}K_1 + \frac{1.125 K_t}{L^2}\\
        \mathcal{V} &=\frac{3EI}{2L^3} + \frac{K_1}{2} + \frac{1.125 K_t}{L^2} \\
        \mathcal{V}_{max} &= \frac{1}{2} \left[ \frac{3EI}{L^3} + K_1 + K_t \left(\frac{1.5}{L}\right)^2 \right]
    \end{aligned}
\end{equation*}
