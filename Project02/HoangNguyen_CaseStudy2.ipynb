## Approach 1.0: Simplify the wing cross section

### Part a: 
Given the mode shape $\phi(x)$ function as following:
$$\phi(x) = 0.5 \left(\frac{x}{L}\right)^2 \left(3 - \frac{x}{L}\right) = \frac{3}{2}\frac{x^2}{L^2} - \frac{1}{2}\frac{x^3}{L^3}$$
$$\phi'(x) = \frac{3x}{L^2} - \frac{3x^2}{2L^3}$$
$$\phi''(x) = \frac{3}{L^2} - \frac{3x}{L^3} = \frac{3}{L^2}\left(1 - \frac{x}{L}\right)$$

At $x= L$, we have 
$$\phi(L) = 0.5(1)^2(3-1) = 1$$
$$\phi'(L) = \frac{3}{L} - \frac{1.5}{L} = \frac{1.5}{L}$$
$$\phi''(x) = \frac{3}{L^2} - \frac{3x}{L^3} = \frac{3}{L^2}\left(1 - \frac{x}{L}\right)$$

* Kinetic energy $\mathcal{T} = \mathcal{T}_{beam} + \mathcal{T}_{tip}$

\begin{equation*}
    \begin{aligned}
        \mathcal{T} &= \mathcal{T}_{beam} + \mathcal{T}_{tip} \\
        \mathcal{T} &= \frac{1}{2} m_w \int_0^L [\phi(x)]^2 dx + \frac{1}{2} M_1 [\phi(L)]^2 + \frac{1}{2} J_{tip} [\phi'(L)]^2\\
        \mathcal{T} &= \frac{1}{2} m_w\int_0^L \left[ \frac{1}{2} \left(\frac{x}{L}\right)^2 \left(3 - \frac{x}{L}\right) \right]^2 dx  +  \frac{1}{2} M_1 (1)^2 +J_{tip} \left(\frac{1.5}{L}\right)^2 \\
        \mathcal{T} &=  m_w \frac{33L}{140} + M_1(1)^2 + J_{tip} \left(\frac{1.5}{L}\right)^2 \\
        \mathcal{T} &= \frac{1}{2} \left[] m_w \frac{33L}{140} + M_1(1)^2 + J_{tip} \left(\frac{1.5}{L}\right)^2 \right]
    \end{aligned}
\end{equation*}

* Potential energy $\mathcal{V} = \mathcal{V}_{beam} + \mathcal{V}_{Springs}$
\begin{equation*}
    \begin{aligned}
        \mathcal{V} &=  \frac{1}{2} EI \int_0^L [\phi''(x)]^2 dx + \frac{1}{2} K_1 [\phi(L)]^2 + \frac{1}{2} K_t [\phi'(L)]^2 \\
        \mathcal{V} &=  \frac{1}{2} EI \int_0^L \left[ \frac{9}{L^4}\left(1 - \frac{x}{L}\right)^2 \right] dx + \frac{1}{2} K_1 (1)^2 + \frac{1}{2} K_t \left(\frac{1.5}{L}\right)^2\\
        \mathcal{V} &= \frac{9EI}{2L^4} \left[ \frac{L}{3} \right]  +\frac{1}{2}K_1 + \frac{1.125 K_t}{L^2}\\
        \mathcal{V} &=\frac{3EI}{2L^3} + \frac{K_1}{2} + \frac{1.125 K_t}{L^2} \\
        \mathcal{V}_{max} &= \frac{1}{2} \left[ \frac{3EI}{L^3} + K_1 + K_t \left(\frac{1.5}{L}\right)^2 \right]
    \end{aligned}
\end{equation*}

From Rayleigh's quotient, we have 
\begin{equation*}
    \begin{aligned}
        \omega^2_n &= \frac{V_{max}}{T_{max}} \\
        \omega_n^2 &= \frac{ \frac{1}{2} \left[] m_w \frac{33L}{140} + M_1(1)^2 + J_{tip} \left(\frac{1.5}{L}\right)^2 \right] }{ \frac{1}{2} \left[ \frac{3EI}{L^3} + K_1 + K_t \left(\frac{1.5}{L}\right)^2 \right] } \\
        \omega_n^2 &=  \frac{ \frac{3EI}{L^3} + K_1 + \frac{2.25 K_t}{L^2} }{ \frac{33 m_w L}{140} + M_1 + \frac{2.25 J_{tip}}{L^2} } 
    \end{aligned}
\end{equation*}

```python
import sympy as sp
import numpy as np

# =========================================================
# 1. Define Symbolic Variables
# =========================================================
x = sp.symbols('x')
L, E, I, m_w = sp.symbols('L E I m_w')  # Beam properties
M_1, J_tip = sp.symbols('M_1 J_{tip}')  # Tip mass properties
K_1, K_t = sp.symbols('K_1 K_t')        # Spring stiffnesses

# =========================================================
# 2. Define Shape Function
# =========================================================
# Given: psi(x) = 0.5 * (x/L)^2 * (3 - x/L)
psi = 0.5 * (x/L)**2 * (3 - x/L)

# Calculate derivatives for Strain Energy
psi_prime = sp.diff(psi, x)        # Slope (theta)
psi_double_prime = sp.diff(psi_prime, x) # Curvature (kappa)

# =========================================================
#   3. Symbolic Integration (Energy Methods)
# =========================================================

# --- Potential Energy (V_max) ---
# Strain Energy of the beam bending
# U_beam = 1/2 * integral( EI * (psi'')^2 dx ) from 0 to L
U_beam = 0.5 * E * I * sp.integrate(psi_double_prime**2, (x, 0, L))

# Potential Energy of Springs at Tip (x=L)
# U_springs = 1/2 * K1 * y(L)^2 + 1/2 * Kt * theta(L)^2
y_tip = psi.subs(x, L)
theta_tip = psi_prime.subs(x, L)
U_springs = 0.5 * K_1 * y_tip**2 + 0.5 * K_t * theta_tip**2

V_max = U_beam + U_springs

# --- Kinetic Energy Reference (T*_max) ---
# Kinetic Energy of Beam Mass
# T_beam = 1/2 * integral( m_w * (psi)^2 dx ) from 0 to L
T_beam = 0.5 * m_w * sp.integrate(psi**2, (x, 0, L))

# Kinetic Energy of Tip Mass and Inertia
# T_tip = 1/2 * M1 * y(L)^2 + 1/2 * J_tip * theta(L)^2
T_tip = 0.5 * M_1 * y_tip**2 + 0.5 * J_tip * theta_tip**2

T_star_max = T_beam + T_tip
```

```python
# =========================================================
# 4. Rayleigh's Quotient Formula
# =========================================================
# omega_n^2 = V_max / T*_max
omega_n_sq = V_max / T_star_max
omega_n = sp.sqrt(omega_n_sq)

print("=== Symbolic Expression for Natural Frequency (omega_n) ===")
omega_n
```

```python
# =========================================================
#               5. Numerical Substitution
# =========================================================

# Geometry Inputs
L_val = 5.0          # m
b_o = 0.70           # m (outer width)
h_o = 0.25           # m (outer height)
t_val = 0.010        # m (wall thickness)

# Material Inputs (Aluminum 6061-T6)
E_val = 69e9         # Pa (N/m^2)
rho = 2700           # kg/m^3

# Tip Mass Inputs
M1_val = 70.0        # kg
tank_radius = 0.13   # m
tank_vol_gal = 16.0  # gallons

# Derived Properties
# Inner dimensions
b_i = b_o - 2*t_val
h_i = h_o - 2*t_val

# Area Moment of Inertia (I) - Hollow Rectangle
I_val = (1/12) * (b_o * h_o**3 - b_i * h_i**3)

# Cross-sectional Area (A)
A_cross = (b_o * h_o) - (b_i * h_i)

# Distributed Mass (m_w)
mw_val = rho * A_cross

# Tip Rotational Inertia (J_tip)
# Cylinder height from volume: V = pi * r^2 * h
vol_m3 = tank_vol_gal * 0.00378541 # Convert gal to m^3
tank_height = vol_m3 / (np.pi * tank_radius**2)
# Inertia about transverse axis: J = m/12 * (3r^2 + h^2)
Jtip_val = (M1_val / 12) * (3 * tank_radius**2 + tank_height**2)

print(f"Computed Properties:")
print(f"  I (m^4):      {I_val:.2e}")
print(f"  m_w (kg/m):   {mw_val:.2f}")
print(f"  J_tip (kgm^2):{Jtip_val:.2f}")

# Substitution Dictionary
# Note: K_1 and K_t are left as symbols or set to 0 if free
params = {
    L: L_val,
    E: E_val,
    I: I_val,
    m_w: mw_val,
    M_1: M1_val,
    J_tip: Jtip_val
}

# Calculate Result
result = omega_n.subs(params)

print("\n--- Final Estimated Natural Frequency (rad/s) ---")
print(f"Equation with unknown stiffnesses K_1 and K_t:")
sp.pprint(result)

# Baseline Estimate (assuming K_1 = 0, K_t = 0 for free cantilever)
baseline_result = result.subs({K_1: 17e6, K_t: 175e6})
print(f"\nBaseline Estimate : {float(baseline_result):.2f} rad/s")
```

### Part b: 

As we decompose beam into two beam-type element, we have 
 * The new length of each element will be $ L_e= 2$ and 
 * There will be three node for node 1 at the tip, node 2 at midpoint length and node 3 at the tip of wing
 * Each node has  vertical displacement $w_e$ and rotation $\theta_e$, we have following global vector 
            $$ q = \begin{Bmatrix} w_1 \\ \theta_1 \\ w_2 \\ \theta_2 \\ w_3 \\ \theta_3\end{Bmatrix}$$
We can solve the finite element problem as following steps:

<b> Step 1 </b>: Constructing element stiffness matricé and element mass matrices

Element Stiffness Matrix ($K^e$):
$$
K^e = \frac{EI}{l^3} \begin{bmatrix}
12 & 6l & -12 & 6l \\
6l & 4l^2 & -6l & 2l^2 \\
-12 & -6l & 12 & -6l \\
6l & 2l^2 & -6l & 4l^2
\end{bmatrix}
$$
Element Consistent Mass Matrix ($M^e$):
$$
M^e = \frac{m\_w l}{420} 
\begin{bmatrix}
156 & 22l & 54 & -13l \\
22l & 4l^2 & 13l & -3l^2 \\
54 & 13l & 156 & -22l \\
-13l & -3l^2 & -22l & 4l^2
\end{bmatrix}
$$

<b> Step 2</b>: Assembling global matrix from each element stiffness
For each element is 4 x 4 matrices, the full assembling global matrix will be 6 x 6 matrices. For each element $L_1 = L_2 = L = L_{beam}/2$, which will be 
* Global stiffness matrix $K_G$:
$$
K_{global} = \frac{EI}{l^3} 
\begin{bmatrix}
12 & 6l & -12 & 6l & 0 & 0 \\
6l & 4l^2 & -6l & 2l^2 & 0 & 0 \\
-12 & -6l & 24 & 0 & -12 & 6l \\
6l & 2l^2 & 0 & 8l^2 & -6l & 2l^2 \\
0 & 0 & -12 & -6l & 12 + \hat{K}_1 & -6l \\
0 & 0 & 6l & 2l^2 & -6l & 4l^2 + \hat{K}_t
\end{bmatrix}
$$
* Global mass matrix $M_G$

$$
M_{global} = \frac{m_w l}{420}
\begin{bmatrix}
156 & 22l & 54 & -13l & 0 & 0 \\
22l & 4l^2 & 13l & -3l^2 & 0 & 0 \\
54 & 13l & 312 & 0 & 54 & -13l \\
-13l & -3l^2 & 0 & 8l^2 & 13l & -3l^2 \\
0 & 0 & 54 & 13l & 156 + \hat{M}_1 & -22l \\
0 & 0 & -13l & -3l^2 & -22l & 4l^2 + \hat{J}
\end{bmatrix}
$$

<b> Step 3 </b>: Apply boundary condition:
As the fixed end is on node 1 for full constrain $w_1 = \theta_1 = 0 $, for that the 
$$ q = \begin{Bmatrix} w_2 \\ \theta_2 \\ w_3 \\ \theta_3\end{Bmatrix}$$
* Reduced stiffness matrix $K_{\text{Reduced}}$:
$$
[K]_{\text{Reduced}} = \frac{EI}{l^3} 
\begin{bmatrix}
24 & 0 & -12 & 6l \\
0 & 8l^2 & -6l & 2l^2 \\
-12 & -6l & 12 + \hat{K}_1 & -6l \\
6l & 2l^2 & -6l & 4l^2 + \hat{K}_t
\end{bmatrix}$$
* Reduced global mass matrix $M_{\text{Reduced}}$ 

$$
[M]_{Reduced} = \frac{m_w l}{420}
\begin{bmatrix}
312 & 0 & 54 & -13l \\
0 & 8l^2 & 13l & -3l^2 \\
54 & 13l & 156 + \hat{M}_1 & -22l \\
-13l & -3l^2 & -22l & 4l^2 + \hat{J}
\end{bmatrix}
$$
* <b> Step 4 </b>: Assembling equation of motion and solve for eigenvalue problem:
For with no damping, we have the equation of motion as following
$$
[M] \ddot{w} + [K] w = 0
$$

Therefore, the eigenvalue problem

$$[K_{global} - \omega^2 M_{global}] \{ \phi \} = 0$$

```python
import sympy as sp
import numpy as np
import scipy.linalg as la

print("=== STEP 1: INPUTS & PARAMETERS ===")
elementNumber = 2
# 1. Define Symbols
E, I, L, mw = sp.symbols('E I L m_w')
# Refined Mesh: 5 Elements
num_elems = elementNumber
l = L / num_elems  # Element length

# Tip Mass/Spring Symbols
M1, Jtip = sp.symbols('M_1 J_{tip}')
K1, Kt = sp.symbols('K_1 K_t')

# 2. Numerical Values
L_val = 5.0
l_val = L_val / num_elems 

# Geometry & Material (Same as before)
b_o, h_o = 0.70, 0.25
t_wall = 0.010
b_i, h_i = b_o - 2*t_wall, h_o - 2*t_wall
E_val = 69e9
rho = 2700

# Calculate derived props
I_val = (1/12) * (b_o * h_o**3 - b_i * h_i**3)
A_cross = (b_o * h_o) - (b_i * h_i)
mw_val = rho * A_cross

# Tip Props
M1_val = 70.0
# Inertia calculation
r_tank = 0.13
vol_m3 = 16.0 * 0.00378541
h_tank = vol_m3 / (np.pi * r_tank**2)
Jtip_val = (M1_val / 12) * (3 * r_tank**2 + h_tank**2)

# Spring Props (Assumed Free)
K1_val = 17e6 #N/m
Kt_val = 175e6 #N/m

print(f"Model: {num_elems} Elements. Element Length l = {l_val} m")     
```

```python
print("=== STEP 2: ASSEMBLING ELEMENTAL MATRICES (SYMBOLIC) ===")
# Coefficients
Ck = (E * I) / l**3
Cm = (mw * l) / 420
    
# These represent the connection between Node i and Node i+1
Ke_base = sp.Matrix([
    [12,   6*l,    -12,   6*l],
    [6*l,  4*l**2, -6*l,  2*l**2],
    [-12,  -6*l,   12,    -6*l],
    [6*l,  2*l**2, -6*l,  4*l**2]
])

Me_base = sp.Matrix([
    [156,   22*l,   54,    -13*l],
    [22*l,  4*l**2, 13*l,  -3*l**2],
    [54,    13*l,   156,   -22*l],
    [-13*l, -3*l**2,-22*l, 4*l**2]
])

Ke = Ck * Ke_base
Me = Cm * Me_base

Ke,Me
 
```

```python
print("=== STEP 3: ASSEMBLING GLOBAL MATRIX (SYMBOLIC) ===")
n_dofs = (num_elems + 1) * 2
K_global = sp.zeros(n_dofs, n_dofs)
M_global = sp.zeros(n_dofs, n_dofs)

for i in range(num_elems): 
    start_idx = 2 * i
    
    for r in range(4):
        for c in range(4):
            K_global[start_idx + r, start_idx + c] += Ke[r, c]
            M_global[start_idx + r, start_idx + c] += Me[r, c]

# Add Tip Dynamics to the Last Node (Node 6)
# The last two indices are n_dofs-2 and n_dofs-1 (Indices 10 and 11)
idx_tip_v = n_dofs - 2
idx_tip_theta = n_dofs - 1

K_global[idx_tip_v, idx_tip_v] += K1
K_global[idx_tip_theta, idx_tip_theta] += Kt

M_global[idx_tip_v, idx_tip_v] += M1
M_global[idx_tip_theta, idx_tip_theta] += Jtip

print(f"Global Assembly Complete. Matrix Size: {n_dofs}x{n_dofs}")
K_global, M_global
```

```python
print("\n=== STEP 4: APPLY BOUNDARY CONDITIONS (REDUCE MATRIX) ===")
# Root (Node 1) is clamped. Remove DOFs 0 and 1.
# We keep indices 2 through 11.
K_reduced = K_global[2:, 2:]
M_reduced = M_global[2:, 2:]

print(f"Boundary conditions applied. Reduced Matrix Size: {K_reduced.shape}")
K_reduced, M_reduced
```

```python
print("=== STEP 5: SUBSTITUTE NUMERICAL VALUES ===")
subs_dict = {
    E: E_val, I: I_val, L: L_val, mw: mw_val,
    M1: M1_val, Jtip: Jtip_val,
    K1: K1_val, Kt: Kt_val
}

# Convert to Numpy arrays
# Note: For large symbolic matrices, element-wise substitution can be slow.
# We use xreplace for better performance or evalf.
K_num = np.array(K_reduced.subs(subs_dict)).astype(np.float64)
M_num = np.array(M_reduced.subs(subs_dict)).astype(np.float64)

print(f"Stiffness Matrix K ({2*num_elems}x{2*num_elems}) calculated.")
print(f"Mass Matrix M ({2*num_elems}x{2*num_elems}) calculated.")
```

```python
print("=== STEP 6: SOLVE EIGENVALUE PROBLEM ===")
vals, vecs = la.eigh(K_num, M_num)

omegas = np.sqrt(vals)
freqs_hz = omegas / (2 * np.pi)

print("=" * 45)
print(f"RESULTS: {num_elems}-ELEMENT FEM NATURAL FREQUENCIES")
print("=" * 45)

for i in range(num_elems*2): 
    print(f"Mode {i+1}: {omegas[i]:.6f} rad/s ({freqs_hz[i]:.4f} Hz)")
    
# Extract the eigenvector for Mode 1
mode1_reduced = vecs[:, 0]

# Reconstruct the full vector (Prepend 0,0 for the clamped root)
    
print("\nMode Shape Vectors (Normalized to Tip Displacement):")
for i in range(len(omegas)): 
    vec = vecs[:, i]
    vec_norm = vec / vec[1] # Normalize to tip displacement               
    print(f"Mode {i+1} Vector: {np.round(vec_norm, 2)}")
```

*<b> Kinetic Energy ($T^*$) </b>: The reference kinetic energy includes the distributed mass of the wing, the discrete mass of the fuel tank ($M_1$), and the rotational inertia of the tank ($J_1$).
$$
T^* = T^*_{beam} + T^*_{tip\_mass} + T^*_{tip\_inertia}
$$
$$
T^* = \frac{1}{2} \int_0^L \rho A_{cs} [w(x)]^2 dx + \frac{1}{2} M_1 [w(L)]^2 + \frac{1}{2} J_1 [w'(L)]^2
$$

```python
import sympy as sp
import numpy as np
import scipy.linalg as la

def solve_element_wing(elementNumber,I_xx,A_cs_num):
    ''' 
    This function solves multi-degree of freedom FEM for a cantilever wing with fix end on
    end and free the other end with tip mass and inertia. The function takes one argument 
    as element number to define the mesh refinement level. The Output are natural frequencies
    in rad/s and Hz along with the mode shape vectors summary. Additionally, the function
    returns omegas, freqs_hz, and vecs for further analysis if needed.
    '''
    # print("=== STEP 1: INPUTS & PARAMETERS ===")
    # 1. Define Symbols
    E, I, L, mw = sp.symbols('E I L m_w') 
    
    num_elems = elementNumber
    
    l = L / num_elems  # Element length
    
    # Tip Mass/Spring Symbols
    M1, Jtip = sp.symbols('M_1 J_{tip}')
    K1, Kt = sp.symbols('K_1 K_t')
    
    # 2. Numerical Values
    L_val = 5.0
    l_val = L_val / num_elems 
    
    # Geometry & Material (Same as before)
    E_val = 69e9
    rho = 2700
    
    # Calculate derived props
    I_val = I_xx_num
    A_cross = A_cs_num
    mw_val = rho * A_cross
    
    # Tip Props
    M1_val = 70.0
    # Inertia calculation
    r_tank = 0.13
    vol_m3 = 16.0 * 0.00378541
    h_tank = vol_m3 / (np.pi * r_tank**2)
    Jtip_val = (M1_val / 12) * (3 * r_tank**2 + h_tank**2)
    
    # Spring Props (Assumed Free)
    K1_val = 0.0
    Kt_val = 0.0

    print(f"Model: {num_elems} Elements. Element Length l = {l_val} m")

    # print("\n=== STEP 2: ASSEMBLING ELEMENTAL MATRICES (SYMBOLIC) ===")
    # Coefficients
    Ck = (E * I) / l**3
    Cm = (mw * l) / 420
     
    # These represent the connection between Node i and Node i+1
    Ke_base = sp.Matrix([
        [12,   6*l,    -12,   6*l],
        [6*l,  4*l**2, -6*l,  2*l**2],
        [-12,  -6*l,   12,    -6*l],
        [6*l,  2*l**2, -6*l,  4*l**2]
    ])
    
    Me_base = sp.Matrix([
        [156,   22*l,   54,    -13*l],
        [22*l,  4*l**2, 13*l,  -3*l**2],
        [54,    13*l,   156,   -22*l],
        [-13*l, -3*l**2,-22*l, 4*l**2]
    ])
    
    Ke = Ck * Ke_base
    Me = Cm * Me_base
     
    # print("\n=== STEP 3: ASSEMBLING GLOBAL MATRIX (SYMBOLIC) ===")
    n_dofs = (num_elems + 1) * 2
    K_global = sp.zeros(n_dofs, n_dofs)
    M_global = sp.zeros(n_dofs, n_dofs)
    
    # Loop to assemble global matrices
    for i in range(num_elems): 
        start_idx = 2 * i
        
        for r in range(4):
            for c in range(4):
                K_global[start_idx + r, start_idx + c] += Ke[r, c]
                M_global[start_idx + r, start_idx + c] += Me[r, c]

     
    idx_tip_v = n_dofs - 2
    idx_tip_theta = n_dofs - 1
    
    K_global[idx_tip_v, idx_tip_v] += K1
    K_global[idx_tip_theta, idx_tip_theta] += Kt
    
    M_global[idx_tip_v, idx_tip_v] += M1
    M_global[idx_tip_theta, idx_tip_theta] += Jtip
    
    print(f"Global Assembly Complete. Matrix Size: {n_dofs}x{n_dofs}")

    # print("\n=== STEP 4: APPLY BOUNDARY CONDITIONS (REDUCE MATRIX) ===")
    # Root (Node 1) is clamped. Remove DOFs 0 and 1. 
    K_reduced = K_global[2:, 2:]
    M_reduced = M_global[2:, 2:]
    
    print(f"Boundary conditions applied. Reduced Matrix Size: {K_reduced.shape}")

    # print("\n=== STEP 5: SUBSTITUTE NUMERICAL VALUES ===")
    subs_dict = {
        E: E_val, I: I_val, L: L_val, mw: mw_val,
        M1: M1_val, Jtip: Jtip_val,
        K1: K1_val, Kt: Kt_val
    }
     
    K_num = np.array(K_reduced.subs(subs_dict)).astype(np.float64)
    M_num = np.array(M_reduced.subs(subs_dict)).astype(np.float64)
    
    print(f"Stiffness Matrix K ({2*num_elems}x{2*num_elems}) calculated.")
    print(f"Mass Matrix M ({2*num_elems}x{2*num_elems}) calculated.")

    # print("\n=== STEP 6: SOLVE EIGENVALUE PROBLEM ===")
    vals, vecs = la.eigh(K_num, M_num)
    
    omegas = np.sqrt(vals)
    freqs_hz = omegas / (2 * np.pi)
    
    print("-" * 40)
    print(f"RESULTS: {num_elems}-ELEMENT FEM NATURAL FREQUENCIES AND TOTAL {num_elems*2} DOFS")
    print("-" * 40)
    
    for i in range(num_elems*2): 
        print(f"Mode {i+1}: {omegas[i]:.6f} rad/s ({freqs_hz[i]:.4f} Hz)")
        
    # Extract the eigenvector for Mode 1
    mode1_reduced = vecs[:, 0]
    
    # Reconstruct the full vector (Prepend 0,0 for the clamped root)
        
    print("\nMode Shape Vectors (Normalized to Tip Displacement):")
    for i in range(len(omegas)): 
        vec = vecs[:, i]
        vec_norm = vec / vec[1] # Normalize to tip displacement
        print(f"Mode {i+1} Vector: {np.round(vec_norm, 2)}")
        
    return omegas, freqs_hz, vecs
        
if __name__ == "__main__":
    solve_element_wing(2,I_xx_num,A_cs_num)    
```

```python
import sympy as sp
'''This cell compute the stiffness and mass matrices for a 2-element cantilever beam with tip masses and inertias at each free node. The root node is fixed, and spring is attached middle node and tip node. The output are global stiffness and mass matrices after applying boundary conditions. Additionally, the force vector is computed considering gravitational effects.
'''
# Define symbols
t = sp.Symbol('t')
l1, l2 = sp.symbols('l_1 l_2')
m1, m2 = sp.symbols('m_1 m_2')
EI1, EI2 = sp.symbols('EI_1 EI_2')
M1, M2 = sp.symbols('M_1 M_2')
K1, K2, kt = sp.symbols('K_1 K_2 k_t')
Ip1, Ip2 = sp.symbols('I_{p1} I_{p2}')
g = sp.Symbol('g')
Ft = sp.Function('F')(t)

# Beam Element Matrices Function
def beam_element(l, EI, m):
    # Stiffness Matrix
    k = (EI / l**3) * sp.Matrix([
        [12, 6*l, -12, 6*l],
        [6*l, 4*l**2, -6*l, 2*l**2],
        [-12, -6*l, 12, -6*l],
        [6*l, 2*l**2, -6*l, 4*l**2]
    ])
    # Mass Matrix (Consistent Mass)
    M_mat = (m * l / 420) * sp.Matrix([
        [156, 22*l, 54, -13*l],
        [22*l, 4*l**2, 13*l, -3*l**2],
        [54, 13*l, 156, -22*l],
        [-13*l, -3*l**2, -22*l, 4*l**2]
    ])
    return k, M_mat

# Generate Element Matrices
# Element 2: Connects Node 3 (fixed) to Node 2. Length l2, properties 2.
k2_el, m2_el = beam_element(l2, EI2, m2)
# Element 1: Connects Node 2 to Node 1. Length l1, properties 1.
k1_el, m1_el = beam_element(l1, EI1, m1)

# Initialize Global Matrices (6x6)
# DOFs: w3, th3, w2, th2, w1, th1
# Indices: 0, 1, 2, 3, 4, 5
K_global = sp.zeros(6, 6)
M_global = sp.zeros(6, 6)

# Assemble Element 2 (Indices 0, 1, 2, 3) -> w3, th3, w2, th2
indices_el2 = [0, 1, 2, 3]
for i in range(4):
    for j in range(4):
        K_global[indices_el2[i], indices_el2[j]] += k2_el[i, j]
        M_global[indices_el2[i], indices_el2[j]] += m2_el[i, j]

# Assemble Element 1 (Indices 2, 3, 4, 5) -> w2, th2, w1, th1
indices_el1 = [2, 3, 4, 5]
for i in range(4):
    for j in range(4):
        K_global[indices_el1[i], indices_el1[j]] += k1_el[i, j]
        M_global[indices_el1[i], indices_el1[j]] += m1_el[i, j]

# Add Discrete Components
# Node 2 (Indices 2, 3): M2, K2, Ip2
# w2 (Index 2): Add K2 to Stiffness, M2 to Mass
K_global[2, 2] += K2
M_global[2, 2] += M2

# th2 (Index 3): Add Ip2 to Mass
M_global[3, 3] += Ip2

# Node 1 (Indices 4, 5): M1, K1, kt, Ip1
# w1 (Index 4): Add K1 to Stiffness, M1 to Mass
K_global[4, 4] += K1
M_global[4, 4] += M1

# th1 (Index 5): Add kt to Stiffness, Ip1 to Mass
K_global[5, 5] += kt
M_global[5, 5] += Ip1

# Apply Boundary Conditions
# Node 3 is fixed: w3=0, th3=0 -> Remove rows/cols 0 and 1
# We want indices 2, 3, 4, 5
active_indices = [2, 3, 4, 5]
K_reduced = K_global[active_indices, active_indices]
M_reduced = M_global[active_indices, active_indices]

# Force Vector
# DOFs: w2, th2, w1, th1
# w2 (Index 0 in reduced): F(t) upwards, -M2*g downwards
# th2 (Index 1 in reduced): 0
# w1 (Index 2 in reduced): -M1*g downwards
# th1 (Index 3 in reduced): 0
F_reduced = sp.Matrix([
    Ft - M2 * g,
    0,
    -M1 * g,
    0
])

# Define displacement vector symbols for display
w2, th2, w1, th1 = sp.symbols('w_2 theta_2 w_1 theta_1')
q = sp.Matrix([w2, th2, w1, th1])
q_ddot = sp.Matrix([sp.Symbol(r'\ddot{w}_2'), sp.Symbol(r'\ddot{\theta}_2'), sp.Symbol(r'\ddot{w}_1'), sp.Symbol(r'\ddot{\theta}_1')])

# Output LaTeX strings for the matrices
# print("Mass Matrix M:")
# print(sp.latex(M_reduced))
# print("\nStiffness Matrix K:")
# print(sp.latex(K_reduced))
# print("\nForce Vector F:")
# print(sp.latex(F_reduced))
```

```python
import sympy as sp
import numpy as np

# ---------------------------------------------------------
# 1. DEFINE PARAMETERS
# ---------------------------------------------------------
# Material: Al 6061-T6
E = 69e9       # Pa (Young's Modulus)
rho = 2700     # kg/m^3 (Density)
nu = 0.33      # Poisson's ratio

# Geometry (Assumed typical fighter wing section for estimation)
# Total Length L = 5.0 m. Let's assume split evenly for the 2 segments.
L_total = 5.0
l1_val = 2.5   # Segment 1 length
l2_val = 2.5   # Segment 2 length

# Cross-section assumptions (Needed to get Area and I)
# Assume a simplified box or airfoil section. 
# For this calculation, we need reasonable estimates for I (Inertia) and A (Area)
# to compute m (mass/len) and EI (stiffness).
# Let's assume an average Chord c=2.0m and Thickness t_avg=0.15m (approx T-33 size)
# Area approx = c * t_avg * struct_factor
# I approx = (1/12) * c * t_avg^3 * struct_factor

# Placeholder values for properties (User should replace with exact geometry if known)
# Area = 0.05 m^2 (approximate effective structural area)
# I = 5e-4 m^4 (approximate effective inertia)
A_approx = 0.05 
I_approx = 5e-4 

m1_val = rho * A_approx  # Mass per unit length
m2_val = m1_val
EI1_val = E * I_approx   # Bending Stiffness
EI2_val = EI1_val

# Discrete Mass/Springs
# M1 (Fuel tank + fuel): 70kg fuel + tank struct (say 30kg) = 100kg
M1_val = 100.0 
# M2 (Payload): Say 200 kg
M2_val = 200.0 

# Tank modeled as cylinder r=0.13m for Inertia
# Ip approx M*r^2/2 (about own axis) + M*dist^2 (if offset). 
# Assuming centered for simplicity or specific Ip provided.
Ip1_val = 0.5 * M1_val * (0.13)**2 
Ip2_val = 10.0 # Arbitrary small inertia for payload M2

# Springs (User inputs)
K1_val = 1e5   # N/m
K2_val = 1e5   # N/m
kt_val = 1e4   # Nm/rad
```

```python
# ---------------------------------------------------------
# 2. SYMBOLIC MATRIX GENERATION (From Part A)
# ---------------------------------------------------------
l1, l2 = sp.symbols('l_1 l_2')
m1, m2 = sp.symbols('m_1 m_2')
EI1, EI2 = sp.symbols('EI_1 EI_2')
M1, M2 = sp.symbols('M_1 M_2')
K1, K2, kt = sp.symbols('K_1 K_2 k_t')
Ip1, Ip2 = sp.symbols('I_{p1} I_{p2}')

# Beam Element Matrices
def beam_matrices(l, EI, m):
    k = (EI / l**3) * sp.Matrix([
        [12, 6*l, -12, 6*l],
        [6*l, 4*l**2, -6*l, 2*l**2],
        [-12, -6*l, 12, -6*l],
        [6*l, 2*l**2, -6*l, 4*l**2]
    ])
    M_mat = (m * l / 420) * sp.Matrix([
        [156, 22*l, 54, -13*l],
        [22*l, 4*l**2, 13*l, -3*l**2],
        [54, 13*l, 156, -22*l],
        [-13*l, -3*l**2, -22*l, 4*l**2]
    ])
    return k, M_mat

k2_el, m2_el = beam_matrices(l2, EI2, m2)
k1_el, m1_el = beam_matrices(l1, EI1, m1)

# Assemble Reduced Global Matrices (Indices 2,3,4,5 correspond to w2, th2, w1, th1)
# (Simplified assembly for brevity - mirrors Part A logic)
K_g = sp.zeros(4, 4)
M_g = sp.zeros(4, 4)

# Element 2 contrib (Node 2 is local node 2)
K_g[0:2, 0:2] += k2_el[2:4, 2:4]
M_g[0:2, 0:2] += m2_el[2:4, 2:4]

# Element 1 contrib (Node 2 is local 1, Node 1 is local 2)
K_g[0:4, 0:4] += k1_el
M_g[0:4, 0:4] += m1_el

# Add Discrete
K_g[0,0] += K2
M_g[0,0] += M2
M_g[1,1] += Ip2
K_g[2,2] += K1
M_g[2,2] += M1
K_g[3,3] += kt
M_g[3,3] += Ip1

# ---------------------------------------------------------
# 3. RAYLEIGH QUOTIENT CALCULATION
# ---------------------------------------------------------

# Define Mode Shape function phi(x) and phi'(x)
x = sp.Symbol('x')
L = l1 + l2
phi = 0.5 * (x/L)**2 * (3 - x/L)
dphi = sp.diff(phi, x)

# Evaluate at Node 2 (x = l2) and Node 1 (x = L)
w2_sym = phi.subs(x, l2)
th2_sym = dphi.subs(x, l2)
w1_sym = phi.subs(x, L)
th1_sym = dphi.subs(x, L)

# Trial Vector q
q_vec = sp.Matrix([w2_sym, th2_sym, w1_sym, th1_sym])

# Symbolic Energy terms
K_eq_sym = (q_vec.T * K_g * q_vec)[0]
M_eq_sym = (q_vec.T * M_g * q_vec)[0]
```

```python
K_eq_sym.simplify()
```

```python
M_eq_sym.simplify() 
```

```python
# ---------------------------------------------------------
# 4. NUMERICAL SUBSTITUTION
# ---------------------------------------------------------
subs_dict = {
    l1: l1_val, l2: l2_val,
    m1: m1_val, m2: m2_val,
    EI1: EI1_val, EI2: EI2_val,
    M1: M1_val, M2: M2_val,
    K1: K1_val, K2: K2_val, kt: kt_val,
    Ip1: Ip1_val, Ip2: Ip2_val
}

K_num = float(K_eq_sym.subs(subs_dict))
M_num = float(M_eq_sym.subs(subs_dict))

omega_n = np.sqrt(K_num / M_num)
fn = omega_n / (2 * np.pi)

print(f"Generalized Stiffness (K_eq): {K_num:.2e}")
print(f"Generalized Mass (M_eq): {M_num:.2e}")
print(f"Fundamental Frequency (omega_n): {omega_n:.2f} rad/s")
print(f"Fundamental Frequency (Hz): {fn:.2f} Hz")
```
