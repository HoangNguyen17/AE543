We have the general solution 

\begin{equation*}
w(x,t) = W(x)T(t)  = \left[ A\cos \left(\frac{\omega x}{c}\right) +B\sin \left( \frac{\omega x}{c} \right) \right] \left[ C \cos (\omega t) + D \sin(\omega t) \right]
\end{equation*}


* At x = 0: 
\begin{equation*}
    P \frac{\partial w}{\partial x} (0,t) = m \frac{\partial^2 w}{\partial t^2}(0,t)
\end{equation*}

From above equation, we have
\begin{equation*}
    \begin{aligned}
    P \frac{\partial w}{\partial x} (0,t) &= m \frac{\partial^2 w}{\partial t^2}(0,t) \\
    P\left[ -\frac{\omega}{c}A\sin\left( \frac{\omega x}{c} \right) + B\frac{\omega}{c}\cos\left( \frac{\omega x}{c} \right) \right] \left[C\cos(\omega t) + D\sin(\omega t)  \right] &=- m \omega^2 \left[ A\cos\left(\frac{\omega x}{c} \right) + B\sin\left(\frac{\omega x}{c}\right) \right] \\
     \left[ C \cos(\omega t) +D\sin(\omega t) \right] \\
    \end{aligned}
\end{equation*}

At x = 0, the above equation can be simplified as
\begin{equation*} 
m\omega^2A + B \left( \frac{P\omega}{c} \right) = 0
\end{equation*}

* At x = L:
\begin{equation*}
    P \frac{\partial w}{\partial x} (L,t) = -k w(L,t)
\end{equation*}

From above equation, we have
\begin{equation*}
    \begin{aligned}
    P \frac{\partial w}{\partial x} (L,t) &= -kw(L,t) \\
    P\left[ -\frac{\omega}{c}A\sin\left( \frac{\omega L}{c} \right) + B\frac{\omega}{c}\cos\left( \frac{\omega L}{c} \right) \right] &=- k \left[ A\cos\left(\frac{\omega L}{c} \right) + B\sin\left(\frac{ \omega L}{c}\right) \right]
    \end{aligned}
\end{equation*}
At x = L, the above equation can be simplified as
\begin{equation*}
    A \left[ -\frac{P\omega}{c}\sin\left( \frac{\omega L}{c} \right) + k \cos\left( \frac{\omega L}{c} \right)\right] + B \left[ \frac{P\omega}{c}\cos \left( \frac{\omega L}{c} \right) +k\sin\left(\frac{\omega L}{c} \right) \right] =0  
\end{equation*}

Therefore, we have 
\begin{equation*}
    \begin{bmatrix}
        m\omega^2 & \frac{P\omega}{c} \\

        \frac{P\omega}{c}\cos\left( \frac{\omega L}{c} \right) + k\sin\left( \frac{\omega L}{c} \right)

    A\\B

We can write as following:

\begin{equation*}

Solve for determinant:

Divide both side for $\cos\left( \frac{\omega L}{c} \right)$, and simplified we have $\tan\alpha = \frac{\sin\alpha}{\cos\alpha}$

### Part a:

We have:
* Kinetic Energy $\mathcal{T} = \mathcal{T}_{bar} + \mathcal{T}_{mass}$ for concentrated mass $M$ at $x =L$: 
        $$ \mathcal{T} = \frac{1}{2}\int_0^L m \dot{u}^2 dx = \frac{1}{2}\rho A \int_0^L \left( \frac{\partial u}{\partial t} \right)^2 dx + \frac{1}{2}M \left[ \frac{\partial u}{\partial t}(L,t) \right]^2 $$
* Potential Energy (assuming neglecting gravitation due to perpendicular to motion): 
        $$\mathcal{V} = \int_0^L \frac{1}{2} EA \left( \frac{\partial u}{\partial x} \right)^2 dx$$ 

Therefore, we have the Lagrangian can be calculate as $ \mathcal{L} = \mathcal{T} - \mathcal{V}$
        $$\mathcal{L} = \frac{1}{2} \int_{0}^{L} \rho A \left( \frac{\partial u}{\partial t} \right)^2 dx + \frac{1}{2} M \left[ \frac{\partial u}{\partial t}(L,t) \right]^2 - \frac{1}{2} \int_{0}^{L} EA \left( \frac{\partial u}{\partial x} \right)^2 dx $$
From Hamilton's Principle $\delta \int_{t_1}^{t_2}(\mathcal{T} - \mathcal{V})dt = 0$,the above equation become
        $$\delta \int_{t_1}^{t_2} \left[ \frac{1}{2} \int_{0}^{L} \rho A \left( \frac{\partial u}{\partial t} \right)^2 dx + \frac{1}{2} M \left( \frac{\partial u}{\partial t}(L,t) \right)^2 - \frac{1}{2} \int_{0}^{L} EA \left( \frac{\partial u}{\partial x} \right)^2 dx \right] dt = 0$$

```python
import numpy as np
import matplotlib.pyplot as plt

def bisection_method(func, a, b, args=(), tol=1e-6, max_iter=100):
    fa = func(a, *args)
    fb = func(b, *args)
    if fa * fb > 0: return None
    for _ in range(max_iter):
        c = (a + b) / 2
        fc = func(c, *args)
        if abs(fc) < tol or (b - a) / 2 < tol: return c
        if fa * fc < 0: b, fb = c, fc
        else: a, fa = c, fc
    return (a + b) / 2

def characteristic_eq_midpoint(lam, mu, kappa):
    # Derived equation:
    # lam*cos(lam) - (lam^2/mu)*sin(lam) - (kappa/2)*sin(lam) + (kappa*lam/mu)*sin^2(lam/2) = 0
    term1 = lam * np.cos(lam)
    term2 = (lam**2 / mu) * np.sin(lam)
    term3 = (kappa / 2) * np.sin(lam)
    term4 = (kappa * lam / mu) * (np.sin(lam/2)**2)
    return term1 - term2 - term3 + term4

mu_values = np.logspace(-2, 2, 200)
kappa_val = 5.0  # Spring at midpoint

roots_orig = [[], [], []] # kappa = 0
roots_mid = [[], [], []]  # kappa = 5.0

# Search bounds
bounds_orig = [(1e-5, 1.57), (3.15, 4.71), (6.29, 7.85)]
bounds_mid = [(1e-5, 3.0), (3.2, 6.0), (6.3, 9.0)]

for mu in mu_values:
    # Original
    for i, (low, high) in enumerate(bounds_orig):
        r = bisection_method(characteristic_eq_midpoint, low, high, args=(mu, 0))
        roots_orig[i].append(r if r else np.nan)
        
    # Midpoint Spring
    for i, (low, high) in enumerate(bounds_mid):
        r = bisection_method(characteristic_eq_midpoint, low, high, args=(mu, kappa_val))
        roots_mid[i].append(r if r else np.nan)

plt.figure(figsize=(10, 6))
plt.plot(mu_values, roots_orig[0], 'b--', label='Mode 1 (Original)')
plt.plot(mu_values, roots_orig[1], 'b--', label='Mode 2 (Original)')
plt.plot(mu_values, roots_orig[2], 'b--', label='Mode 3 (Original)')

plt.plot(mu_values, roots_mid[0], 'g-', linewidth=2, label=f'Mode 1 (Spring at L/2, $\kappa$={kappa_val})')
plt.plot(mu_values, roots_mid[1], 'g-', linewidth=2, label=f'Mode 2 (Spring at L/2, $\kappa$={kappa_val})')
plt.plot(mu_values, roots_mid[2], 'g-', linewidth=2, label=f'Mode 3 (Spring at L/2, $\kappa$={kappa_val})')

plt.xscale('log')
plt.xlabel(r'Mass Ratio $\mu = \frac{\rho A L}{M}$')
plt.ylabel(r'Frequency Parameter $\lambda$')
plt.title('Natural Frequencies: Original vs. Spring at Midpoint')
plt.legend()
plt.grid(True, which="both", alpha=0.3)
plt.tight_layout()
plt.savefig('midpoint_spring_plot.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def bisection_method(func, a, b, args=(), tol=1e-6, max_iter=100):
    fa = func(a, *args)
    fb = func(b, *args)
    if fa * fb > 0: return None
    for _ in range(max_iter):
        c = (a + b) / 2
        fc = func(c, *args)
        if abs(fc) < tol or (b - a) / 2 < tol: return c
        if fa * fc < 0: b, fb = c, fc
        else: a, fa = c, fc
    return (a + b) / 2

# Characteristic equation with stiffness kappa
# cot(lam) = lam/mu - kappa/lam
# Continuous form: lam * cos(lam) - (lam^2/mu - kappa) * sin(lam) = 0
def characteristic_eq_stiff(lam, mu, kappa):
    return lam * np.cos(lam) - (lam**2 / mu - kappa) * np.sin(lam)

mu_values = np.logspace(-2, 2, 200)
kappa_val = 2.0  # Dimensionless stiffness k_a * L / EA

# Storage
roots_orig = [[], [], []] # kappa = 0
roots_stiff = [[], [], []] # kappa = 2.0

# Bounds need care as roots shift with stiffness
# For kappa=0 (original), roots are in (0, pi/2), (pi, 3pi/2)...
# For kappa > 0, roots shift upwards. 
# Mode 1 can shift past pi/2 if stiffness is high enough?
# Let's use wide search windows.
bounds_orig = [(1e-5, 1.57), (3.15, 4.71), (6.29, 7.85)]
bounds_stiff = [(1e-5, 3.0), (3.2, 6.0), (6.3, 9.0)] # Wider search for stiffened

for mu in mu_values:
    # Original (kappa=0)
    for i, (low, high) in enumerate(bounds_orig):
        r = bisection_method(characteristic_eq_stiff, low, high, args=(mu, 0))
        if r: roots_orig[i].append(r)
        else: roots_orig[i].append(np.nan) # Fallback
        
    # Stiffened (kappa=2.0)
    for i, (low, high) in enumerate(bounds_stiff):
        # Dynamic adjustment of bounds if needed, but fixed wide usually works for simple cases
        r = bisection_method(characteristic_eq_stiff, low, high, args=(mu, kappa_val))
        if r: roots_stiff[i].append(r)
        else: roots_stiff[i].append(np.nan)

plt.figure(figsize=(10, 6))

# Plot Original
plt.plot(mu_values, roots_orig[0], 'b--', label='Mode 1 (Original)')
plt.plot(mu_values, roots_orig[1], 'b--', label='Mode 2 (Original)')
plt.plot(mu_values, roots_orig[2], 'b--', label='Mode 3 (Original)')

# Plot Stiffened
plt.plot(mu_values, roots_stiff[0], 'r-', linewidth=2, label=f'Mode 1 (Added Spring $\kappa={kappa_val}$)')
plt.plot(mu_values, roots_stiff[1], 'r-', linewidth=2, label=f'Mode 2 (Added Spring $\kappa={kappa_val}$)')
plt.plot(mu_values, roots_stiff[2], 'r-', linewidth=2, label=f'Mode 3 (Added Spring $\kappa={kappa_val}$)')

plt.xscale('log')
plt.xlabel(r'Mass Ratio $\mu = \frac{\rho A L}{M}$')
plt.ylabel(r'Frequency Parameter $\lambda$')
plt.title('Natural Frequencies: Original vs. Stiffened Tip')
plt.legend()
plt.grid(True, which="both", alpha=0.3)
plt.tight_layout()
plt.savefig('stiffened_frequency_plot.png')
```

### Part a:

Kinetic Energy ($T$):The kinetic energy of the shaft is due to the rotational velocity ($\frac{\partial \theta}{\partial t}$ or $\dot{\theta}$) of the mass.
$$T = \frac{1}{2} \int_0^L I_{shaft} \left( \frac{\partial \theta}{\partial t} \right)^2 dx$$
Potential Energy ($V$):The potential energy is the elastic strain energy stored in the shaft due to twisting ($\frac{\partial \theta}{\partial x}$ or $\theta'$).
$$V = \frac{1}{2} \int_0^L GJ \left( \frac{\partial \theta}{\partial x} \right)^2 dx$$

Hamilton's Principle states that the variation of the time integral of the Lagrangian ($L = T - V$) is zero between two time instants $t_1$ and $t_2$:
$$\delta \int_{t_1}^{t_2} (T - V) dt = 0$$
Substituting our energy expressions:
$$\delta \int_{t_1}^{t_2} \left[ \frac{1}{2} \int_0^L I_{shaft} \left( \frac{\partial \theta}{\partial t} \right)^2 dx - \frac{1}{2} \int_0^L GJ \left( \frac{\partial \theta}{\partial x} \right)^2 dx \right] dt = 0$$
