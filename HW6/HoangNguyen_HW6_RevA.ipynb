### Part a:

We have:
* Kinetic Energy $\mathcal{T} = \mathcal{T}_{bar} + \mathcal{T}_{mass}$ for concentrated mass $M$ at $x =L$: 
        $$ \mathcal{T} = \frac{1}{2}\int_0^L m \dot{u}^2 dx = \frac{1}{2}\rho A \int_0^L \left( \frac{\partial u}{\partial t} \right)^2 dx + \frac{1}{2}M \left[ \frac{\partial u}{\partial t}(L,t) \right]^2 $$
* Potential Energy (assuming neglecting gravitation due to perpendicular to motion): 
        $$\mathcal{V} = \int_0^L \frac{1}{2} EA \left( \frac{\partial u}{\partial x} \right)^2 dx$$ 

Therefore, we have the Lagrangian can be calculate as $ \mathcal{L} = \mathcal{T} - \mathcal{V}$
        $$\mathcal{L} = \frac{1}{2} \int_{0}^{L} \rho A \left( \frac{\partial u}{\partial t} \right)^2 dx + \frac{1}{2} M \left[ \frac{\partial u}{\partial t}(L,t) \right]^2 - \frac{1}{2} \int_{0}^{L} EA \left( \frac{\partial u}{\partial x} \right)^2 dx $$
From Hamilton's Principle $\delta \int_{t_1}^{t_2}(\mathcal{T} - \mathcal{V})dt = 0$,the above equation become
        $$\delta \int_{t_1}^{t_2} \left[ \frac{1}{2} \int_{0}^{L} \rho A \left( \frac{\partial u}{\partial t} \right)^2 dx + \frac{1}{2} M \left( \frac{\partial u}{\partial t}(L,t) \right)^2 - \frac{1}{2} \int_{0}^{L} EA \left( \frac{\partial u}{\partial x} \right)^2 dx \right] dt = 0$$

we can intergrate seperate each term as
* **Term 1: variation of kinetic energy at the bar**

$$\delta \int_{t_1}^{t_2} \frac{1}{2} \int_{0}^{L} \rho A \left( \frac{\partial u}{\partial t} \right)^2 dx \, dt = \int_{t_1}^{t_2} \int_{0}^{L} \rho A \frac{\partial u}{\partial t} \delta \left( \frac{\partial u}{\partial t} \right) dx \, dt$$
Integrate by part for $\int u dv = uv - \int v du$: 
$$\int_{t_1}^{t_2} \int_{0}^{L} \rho A \frac{\partial u}{\partial t} \delta \left( \frac{\partial u}{\partial t} \right) dx dt =  \int_{0}^{L} \left[ \rho A \frac{\partial u}{\partial t} \delta u \right]_{t_1}^{t_2} dx - \int_{t_1}^{t_2} \int_{0}^{L} \rho A \frac{\partial^2 u}{\partial t^2} \delta u \, dx \, dt $$

with $\delta u = 0$ at $t_1$ and $t_2$, we have  
$$\int_{t_1}^{t_2} \int_{0}^{L} \rho A \frac{\partial u}{\partial t} \delta \left( \frac{\partial u}{\partial t} \right) dx dt = - \int_{t_1}^{t_2} \int_{0}^{L} \rho A \frac{\partial^2 u}{\partial t^2} \delta u \, dx \, dt $$

* **Term 2: Variation of kinetic energy at the mass**

$$\delta \int_{t_1}^{t_2} \frac{1}{2} M \left( \frac{\partial u}{\partial t}(L,t) \right)^2 dt = \int_{t_1}^{t_2} M \frac{\partial u}{\partial t}(L,t) \delta \left( \frac{\partial u}{\partial t}(L,t) \right) dt$$
Integrate by part:

$$\int_{t_1}^{t_2} M \frac{\partial u}{\partial t}(L,t) \delta \left( \frac{\partial u}{\partial t}(L,t) \right) dt= \left[ M \frac{\partial u}{\partial t}(L,t) \delta u(L,t) \right]_{t_1}^{t_2} - \int_{t_1}^{t_2} M \frac{\partial^2 u}{\partial t^2}(L,t) \delta u(L,t) \, dt$$
with $\delta u = 0$ at $t_1$ and $t_2$, we have 
$$\int_{t_1}^{t_2} M \frac{\partial u}{\partial t}(L,t) \delta \left( \frac{\partial u}{\partial t}(L,t) \right) dt= - \int_{t_1}^{t_2} M \frac{\partial^2 u}{\partial t^2}(L,t) \delta u(L,t) \, dt$$

* **Term 3: Variation of potential energy**

$$\delta \int_{t_1}^{t_2} -\frac{1}{2} \int_{0}^{L} EA \left( \frac{\partial u}{\partial x} \right)^2 dx \, dt = -\int_{t_1}^{t_2} \int_{0}^{L} EA \frac{\partial u}{\partial x} \delta \left( \frac{\partial u}{\partial x} \right) dx \, dt$$
Intergrate by part:

$$-\int_{t_1}^{t_2} \int_{0}^{L} EA \frac{\partial u}{\partial x} \delta \left( \frac{\partial u}{\partial x} \right) dx  =  -\int_{t_1}^{t_2} \left( \left[ EA \frac{\partial u}{\partial x} \delta u \right]_{0}^{L} - \int_{0}^{L} \frac{\partial}{\partial x} \left( EA \frac{\partial u}{\partial x} \right) \delta u \, dx \right) dt$$

From all compute term above we have:

\begin{equation*}
        \begin{aligned}
        \delta \int_{t_1}^{t_2} \left[ \frac{1}{2} \int_{0}^{L} \rho A \left( \frac{\partial u}{\partial t} \right)^2 dx + \frac{1}{2} M \left( \frac{\partial u}{\partial t}(L,t) \right)^2 - \frac{1}{2} \int_{0}^{L} EA \left( \frac{\partial u}{\partial x} \right)^2 dx \right] dt &= 0 \\
         - \int_{t_1}^{t_2} \int_{0}^{L} \rho A \frac{\partial^2 u}{\partial t^2} \delta u \, dx \, dt   - \int_{t_1}^{t_2} M \frac{\partial^2 u}{\partial t^2}(L,t) \delta u(L,t) \, dt   -\int_{t_1}^{t_2} \left( \left[ EA \frac{\partial u}{\partial x} \delta u \right]_{0}^{L} - \int_{0}^{L} \frac{\partial}{\partial x} \left( EA \frac{\partial u}{\partial x} \right) \delta u \, dx \right) dt&=0\\
         \text{Simplify above and solve the space integrate over the domain $0<x<L$ and the boundary term $x = 0$ and $x =L$} \\
         \int_{t_1}^{t_2} \left\{ \underbrace{\int_{0}^{L} \left[ -\rho A \frac{\partial^2 u}{\partial t^2} + \frac{\partial}{\partial x} \left( EA \frac{\partial u}{\partial x} \right) \right] \delta u \, dx}_{\text{Differential Equation}}
- \underbrace{\left[ M \frac{\partial^2 u}{\partial t^2}(L,t) + EA \frac{\partial u}{\partial x}(L,t) \right] \delta u(L,t)}_{\text{Boundary Condition at } x=L}
+ \underbrace{\left[ EA \frac{\partial u}{\partial x}(0,t) \right] \delta u(0,t)}_{\text{Boundary Condition at } x=0} \right\} dt = 0
        \end{aligned}
\end{equation*}

Therefore, we can solve for the equation of motion as
$$-\rho A \frac{\partial^2 u}{\partial t^2} + \frac{\partial}{\partial x} \left( EA \frac{\partial u}{\partial x} \right) = 0$$
For, $ -\rho A = M$, the above equation of motion is
$$\boxed{\mathbf{M \frac{\partial^2 u}{\partial t^2}(x,t) - EA \frac{\partial^2 u}{\partial x^2}(x,t) = 0}}$$

* Apply boundary condition at $x = 0$, we have $$u(0,t)=0$$
* Apply boundary condition at $x = L$, we have 
$$-\left[ M \frac{\partial^2 u}{\partial t^2}(L,t) + EA \frac{\partial u}{\partial x}(L,t) \right] = 0$$
$$ EA \frac{\partial^2 u}{\partial x^2}(L,t) = - M \frac{\partial^2 u}{\partial t^2}(L,t)  $$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
def get_roots(mu_values):
    roots = np.zeros((len(mu_values), 3))
    # Define the function for the characteristic equation:
    # lambda * tan(lambda) = mu  =>  lambda * sin(lambda) - mu * cos(lambda) = 0
    # We use the continuous form to avoid singularities at tan(lambda) asymptotes.
    def char_eq(lam, mu):
        return lam * np.sin(lam) - mu * np.cos(lam)

    # Bracketing intervals for the first three roots
    # These intervals are valid for mu > 0.
    # Root 1 is in (0, pi/2)
    # Root 2 is in (pi, 3pi/2)
    # Root 3 is in (2pi, 5pi/2)
    intervals = [
        (1e-6, np.pi/2 - 1e-6),
        (np.pi + 1e-6, 3*np.pi/2 - 1e-6),
        (2*np.pi + 1e-6, 5*np.pi/2 - 1e-6)
    ]

    for i, mu in enumerate(mu_values):
        current_roots = []
        for (a, b) in intervals:
            # Bisection method implementation using numpy
            low = a
            high = b
            for _ in range(50): # 50 iterations is plenty for plotting precision
                mid = (low + high) / 2
                f_mid = char_eq(mid, mu)
                f_low = char_eq(low, mu)
                
                if f_mid * f_low < 0:
                    high = mid
                else:
                    low = mid
            current_roots.append((low + high) / 2)
        roots[i, :] = current_roots
    return roots

# Define the range of mass ratio rho*A*L / M
# Using logspace to cover regimes where mass is dominant vs bar is dominant
mu = np.logspace(-2, 2, 200)

lambdas = get_roots(mu)

plt.figure(figsize=(10, 6))
plt.plot(mu, lambdas[:, 0], label='Mode 1')
plt.plot(mu, lambdas[:, 1], label='Mode 2')
plt.plot(mu, lambdas[:, 2], label='Mode 3')
original_root=[lambdas[:, 0], lambdas[:, 1], lambdas[:, 2]]
# Add asymptotic limits for visual reference
# Fixed-Free limits (M -> 0 => mu -> infinity): pi/2, 3pi/2, 5pi/2
plt.axhline(np.pi/2, color='k', linestyle='--', alpha=1, label=r'Fixed-Free Limits ($\mu \to \infty$)')
plt.axhline(3*np.pi/2, color='k', linestyle='--', alpha=1)
plt.axhline(5*np.pi/2, color='k', linestyle='--', alpha=1)

# Fixed-Fixed limits (M -> infinity => mu -> 0): 0, pi, 2pi
plt.axhline(0, color='r', linestyle=':', alpha=1, label=r'Fixed-Fixed Limits ($\mu \to 0$)')
plt.axhline(np.pi, color='r', linestyle=':', alpha=1)
plt.axhline(2*np.pi, color='r', linestyle=':', alpha=1)

# plt.xscale('log')
plt.xlabel(r'Mass Ratio $\mu = \frac{\rho A L}{M}$')
plt.ylabel(r'Frequency Parameter $\lambda = \omega L \sqrt{\rho/E}$')
plt.title(r'First Three Natural Frequencies vs Mass Ratio')
plt.grid(True, which="both", alpha=0.3)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('natural_frequencies_part_b.png')
```

#### <i>Section 1 <i>: Derive equation of motion


When we add a spring $k_a$ and a damper $c_a$ at $x= \frac{L}{2}$, we have
* Kinetic energy: 
$$\mathcal{T} = \mathcal{T}_{rod} + \mathcal{T}_{mass}$$
$$\mathcal{T} = \int_{0}^{L} \frac{1}{2} \rho A \left( \frac{\partial u}{\partial t} \right)^2 dx + \frac{1}{2} M \left[ \frac{\partial u}{\partial t}(L,t) \right]^2$$
$$\delta \mathcal{T} = -\int_{0}^{L} \rho A \frac{\partial^2 u}{\partial t^2} \delta u \, dx - M \frac{\partial^2 u}{\partial t^2}(L,t) \delta u(L,t)$$
* Potential energy:

$$ \mathcal{V} = \mathcal{V}_{\text{Rod Strain}} + \mathcal{V}_{\text{Spring at $L/2$}} $$
$$\mathcal{V} = \int_{0}^{L} \frac{1}{2} EA \left( \frac{\partial u}{\partial x} \right)^2 dx + \frac{1}{2} k_a \left[ u\left(\frac{L}{2},t\right) \right]^2$$
$$\delta \mathcal{V} = \int_{0}^{L} \left( -EA \frac{\partial^2 u}{\partial x^2} \right) \delta u \, dx + \left[ EA \frac{\partial u}{\partial x} \delta u \right]_{0}^{L/2} + \left[ EA \frac{\partial u}{\partial x} \delta u \right]_{L/2}^{L} + k_a u\left(\frac{L}{2}\right) \delta u\left(\frac{L}{2}\right)$$
As having con-conservative damper force $F_d = c_a \frac{\partial u}{\partial t} $ at $ x = \frac{L}{2}$

$$\delta W_{nc} = -c_a \frac{\partial u}{\partial t}\left(\frac{L}{2},t\right) \delta u\left(\frac{L}{2},t\right)$$

From Hamilton's principle:
$$\int_{t_1}^{t_2} (\delta \mathcal{T} - \delta \mathcal{V} + \delta W_{nc}) \, dt = 0$$

$$\int_{t_1}^{t_2} \left[ \int_{0}^{L} \left( -\rho A \frac{\partial^2 u}{\partial t^2} + EA \frac{\partial^2 u}{\partial x^2} \right) \delta u \, dx + \left. \left( EA \frac{\partial u}{\partial x} \delta u \right) \right|_{0}^{L} - M \frac{\partial^2 u}{\partial t^2}(L,t) \delta u(L,t) - k_a u\left(\frac{L}{2}\right) \delta u\left(\frac{L}{2}\right) \right] dt = 0$$
Therefore, equation of motion is: 

$$-\rho A \frac{\partial^2 u}{\partial t^2} + EA \frac{\partial^2 u}{\partial x^2} = 0 \quad \Rightarrow \quad \rho A \frac{\partial^2 u}{\partial t^2} = EA \frac{\partial^2 u}{\partial x^2}$$

With boundary condition 
* at $x = 0$: 
$$ u(0,t) = 0$$

* at $x = L$: 
$$\left( EA \frac{\partial u}{\partial x}(L,t) - M \frac{\partial^2 u}{\partial t^2}(L,t) \right) \delta u(L,t) = 0$$
$$EA \frac{\partial u}{\partial x}(L,t) = M \frac{\partial^2 u}{\partial t^2}(L,t)$$

* at $x = \frac{L}{2}$:
$$\left( EA \frac{\partial u}{\partial x}(L,t) - M \frac{\partial^2 u}{\partial t^2}(L,t) \right) \delta u(L,t) = 0$$
$$EA \frac{\partial u}{\partial x}(L,t) = M \frac{\partial^2 u}{\partial t^2}(L,t)$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Custom bisection method for root finding
def bisection_method(func, a, b, args=(), tol=1e-6, max_iter=100):
    fa = func(a, *args)
    fb = func(b, *args)
    if fa * fb > 0: return None
    for _ in range(max_iter):
        c = (a + b) / 2
        fc = func(c, *args)
        if abs(fc) < tol or (b - a) / 2 < tol: return c
        if fa * fc < 0: b, fb = c, fc
        else: a, fa = c, fc
    return (a + b) / 2

# Characteristic eq for rod with tip mass M and spring k_a at L/2 (undamped)
# Eq: lam*cos(lam) - (lam^2/mu)*sin(lam) - (kappa/2)*sin(lam) + (kappa*lam/mu)*sin^2(lam/2) = 0
def characteristic_eq_midpoint(lam, mu, kappa):
    term1 = lam * np.cos(lam)
    term2 = (lam**2 / mu) * np.sin(lam)
    term3 = (kappa / 2) * np.sin(lam)
    term4 = (kappa * lam / mu) * (np.sin(lam/2)**2)
    return term1 - term2 - term3 + term4

mu_values = np.logspace(-2, 2, 200)
kappa_val = 1.5  # Stiffness ratio for the spring at L/2

roots_mid = [[], [], []]  # Stores the roots for the modified system

# Search bounds are slightly adjusted due to the stiffness
bounds_mid = [(1e-5, 3.0), (3.2, 6.0), (6.3, 9.0)]

for mu in mu_values:
    # Midpoint Spring
    for i, (low, high) in enumerate(bounds_mid):
        r = bisection_method(characteristic_eq_midpoint, low, high, args=(mu, kappa_val))
        roots_mid[i].append(r if r else np.nan)

# Plotting the modified system only
plt.figure(figsize=(10, 6))
plt.plot(mu_values, roots_mid[0], 'g-', linewidth=2, label=r'1st Mode ($\lambda_1$)')
plt.plot(mu_values, roots_mid[1], 'r-', linewidth=2, label=r'2nd Mode ($\lambda_2$)')
plt.plot(mu_values, roots_mid[2], 'b-', linewidth=2, label=r'3rd Mode ($\lambda_3$)')
spring_roots = [roots_mid[0], roots_mid[1], roots_mid[2]]
# Add asymptotic limits (Fixed-Free limits are relevant as mu -> inf)
plt.axhline(np.pi/2, color='k', linestyle=':', alpha=1.0)
plt.axhline(3*np.pi/2, color='k', linestyle=':', alpha=1.0)
plt.axhline(5*np.pi/2, color='k', linestyle=':', alpha=1.0, label=r'Asymptotic Fixed-Free Limits')

plt.xscale('log')
plt.xlabel(r'Mass Ratio $\mu = \frac{\rho A L}{M}$ (Log Scale)')
plt.ylabel(r'Frequency Parameter $\lambda = \omega L \sqrt{\rho/E}$')
plt.title(rf'Natural Frequencies of Rod with Spring at $L/2$ ($\kappa={kappa_val}$)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, which="both", alpha=0.3)
plt.tight_layout()
plt.savefig('midpoint_spring_only_plot.png')
```

## Problem 3

### Part a:

Kinetic Energy ($T$):The kinetic energy of the shaft is due to the rotational velocity ($\frac{\partial \theta}{\partial t}$ or $\dot{\theta}$) of the mass.
$$T = \frac{1}{2} \int_0^L I_{shaft} \left( \frac{\partial \theta}{\partial t} \right)^2 dx$$
Potential Energy ($V$):The potential energy is the elastic strain energy stored in the shaft due to twisting ($\frac{\partial \theta}{\partial x}$ or $\theta'$).
$$V = \frac{1}{2} \int_0^L GJ \left( \frac{\partial \theta}{\partial x} \right)^2 dx$$

Hamilton's Principle states that the variation of the time integral of the Lagrangian ($L = T - V$) is zero between two time instants $t_1$ and $t_2$:

$$\delta \int_{t_1}^{t_2} (T - V) dt = 0$$

Substituting our energy expressions:

$$\delta \int_{t_1}^{t_2} \left[ \frac{1}{2} \int_0^L I_{shaft} \left( \frac{\partial \theta}{\partial t} \right)^2 dx - \frac{1}{2} \int_0^L GJ \left( \frac{\partial \theta}{\partial x} \right)^2 dx \right] dt = 0$$

$$\int_{t_1}^{t_2} \int_0^L \left[ I_{shaft} \frac{\partial \theta}{\partial t} \delta \left( \frac{\partial \theta}{\partial t} \right) - GJ \frac{\partial \theta}{\partial x} \delta \left( \frac{\partial \theta}{\partial x} \right) \right] dx dt = 0$$
For $\delta(\frac{\partial \theta}{\partial t}) = \frac{\partial}{\partial t}(\delta \theta)$ and $\delta(\frac{\partial \theta}{\partial x}) = \frac{\partial}{\partial x}(\delta \theta)$:

$$\int_{t_1}^{t_2} \int_0^L \left[ I_{shaft} \frac{\partial \theta}{\partial t} \frac{\partial}{\partial t} (\delta \theta) - GJ \frac{\partial \theta}{\partial x} \frac{\partial}{\partial x} (\delta \theta) \right] dx dt = 0$$

* **Term 1**
$$\int_{t_1}^{t_2} I_{shaft} \frac{\partial \theta}{\partial t} \frac{\partial}{\partial t} (\delta \theta) dt = \left[ I_{shaft} \frac{\partial \theta}{\partial t} \delta \theta \right]_{t_1}^{t_2} - \int_{t_1}^{t_2} I_{shaft} \frac{\partial^2 \theta}{\partial t^2} \delta \theta dt$$
as $\delta \theta(t_1) = \delta \theta(t_2) = 0$, we have
$$\int_{t_1}^{t_2} I_{shaft} \frac{\partial \theta}{\partial t} \frac{\partial}{\partial t} (\delta \theta) dt = - \int_{t_1}^{t_2} I_{shaft} \frac{\partial^2 \theta}{\partial t^2} \delta \theta dt$$

* **Term 2**
$$\int_0^L GJ \frac{\partial \theta}{\partial x} \frac{\partial}{\partial x} (\delta \theta) dx = \left[ GJ \frac{\partial \theta}{\partial x} \delta \theta \right]_0^L - \int_0^L \frac{\partial}{\partial x} \left( GJ \frac{\partial \theta}{\partial x} \right) \delta \theta dx$$

Substitute:
$$\int_{t_1}^{t_2} \int_0^L \left[ I_{shaft} \frac{\partial \theta}{\partial t} \frac{\partial}{\partial t} (\delta \theta) - GJ \frac{\partial \theta}{\partial x} \frac{\partial}{\partial x} (\delta \theta) \right] dx dt = 0$$
$$\int_{t_1}^{t_2} \int_0^L \left[ -I_{shaft} \frac{\partial^2 \theta}{\partial t^2} + \frac{\partial}{\partial x} \left( GJ \frac{\partial \theta}{\partial x} \right) \right] \delta \theta dx dt - \int_{t_1}^{t_2} \left[ GJ \frac{\partial \theta}{\partial x} \delta \theta \right]_0^L dt = 0$$

Therefore, the equation of motion is 

$$\boxed{I_{shaft} \frac{\partial^2 \theta}{\partial t^2} - GJ \frac{\partial^2 \theta}{\partial x^2} = 0}$$
And the boundary condition term is 
$$- \int_{t_1}^{t_2} \left[ GJ \frac{\partial \theta}{\partial x} \delta \theta \right]_0^L dt = 0$$
$$\left( GJ \frac{\partial \theta}{\partial x} \delta \theta \right)\Bigg|_{x=L} - \left( GJ \frac{\partial \theta}{\partial x} \delta \theta \right)\Bigg|_{x=0} = 0$$
* At $x = 0$, the shaft is fixed and rotation is zero,  $\boxed{\theta(0,t) = 0}$
* At $x = L$, The shaft is free where $\delta \theta(L,t) \neq 0$, so $GJ \frac{\partial \theta}{\partial x} \bigg|_{x=L} = 0$. For that we have 
$$GJ \frac{\partial \theta}{\partial x} \bigg|_{x=L} = 0 \Leftrightarrow \boxed{ \frac{\partial \theta}{\partial x}(L,t) = 0}$$

#### Section 1: Derive equation of motion
Now we add a round disk at location $x = L$ with polar mass moment intertia $I_{gear}$, the equation can be
* Kinetic energy:
$$\mathcal{T} = \underbrace{\frac{1}{2} \int_0^L I_{shaft} \left( \frac{\partial \theta}{\partial t} \right)^2 dx}_{\text{Shaft}} + \underbrace{\frac{1}{2} I_{gear} \left( \frac{\partial \theta(L,t)}{\partial t} \right)^2}_{\text{Gear at } x=L}$$
* Potential energy:
$$\mathcal{V} = \frac{1}{2} \int_0^L GJ \left( \frac{\partial \theta}{\partial x} \right)^2 dx$$
From Hamilton's principle

$$\delta \int_{t_1}^{t_2} (T - V) dt = 0$$
$$\delta \int_{t_1}^{t_2} \left[ \frac{1}{2} \int_0^L I_{shaft} \dot{\theta}^2 dx + \frac{1}{2} I_{gear} \dot{\theta}(L)^2 - \frac{1}{2} \int_0^L GJ (\theta')^2 dx \right] dt = 0$$
$$\int_{t_1}^{t_2} \left[ \int_0^L (I_{shaft} \dot{\theta} \delta \dot{\theta} - GJ \theta' \delta \theta') dx + I_{gear} \dot{\theta}(L) \delta \dot{\theta}(L) \right] dt = 0$$

* Term 1:

$$\int_{t_1}^{t_2} \int_0^L I_{shaft} \dot{\theta} \delta \dot{\theta} dx dt = - \int_{t_1}^{t_2} \int_0^L I_{shaft} \ddot{\theta} \delta \theta dx dt$$

(Boundary terms at $t_1, t_2$ vanish).

* Term 2:
$$-\int_{t_1}^{t_2} \int_0^L GJ \theta' \delta \theta' dx dt = \int_{t_1}^{t_2} \int_0^L GJ \theta'' \delta \theta dx dt - \int_{t_1}^{t_2} \left[ GJ \theta' \delta \theta \right]_0^L dt$$

* Term 3:

$$\int_{t_1}^{t_2} I_{gear} \dot{\theta}(L) \delta \dot{\theta}(L) dt = \left[ I_{gear} \dot{\theta}(L) \delta \theta(L) \right]_{t_1}^{t_2} - \int_{t_1}^{t_2} I_{gear} \ddot{\theta}(L) \delta \theta(L) dt$$

As $\delta \theta = 0$ over the boundary $t_1$ and $t_2$

$$ \int_{t_1}^{t_2} I_{gear} \dot{\theta}(L) \delta \dot{\theta}(L) dt = - \int_{t_1}^{t_2} I_{gear} \ddot{\theta}(L) \delta \theta L) dt $$

From all terms, we have the equation of motion as follow
$$ \int_{t_1}^{t_2} \int_0^L \left( -I_{shaft} \ddot{\theta} + GJ \theta'' \right) \delta \theta dx dt $$
$$\boxed{I_{shaft} \frac{\partial^2 \theta}{\partial t^2} - GJ \frac{\partial^2 \theta}{\partial x^2} = 0}$$
Boundary Term at $x=L$, At the free end, $\delta \theta(L)$ is arbitrary (not zero). Therefore, the coefficient in the boundary integral must be zero::

$$\int_{t_1}^{t_2} \left( - \underbrace{GJ \theta'(L) \delta \theta(L)}_{\text{From Strain Energy}} - \underbrace{I_{gear} \ddot{\theta}(L) \delta \theta(L)}_{\text{From Gear Kinetic Energy}} \right) dt$$
$$\boxed{GJ \frac{\partial \theta}{\partial x}(L,t) + I_{gear} \frac{\partial^2 \theta}{\partial t^2}(L,t) = 0}$$

Boundary Term at $x=0$: $\int_{t_1}^{t_2} GJ \theta'(0) \delta \theta(0) dt$ at the fixed end, $\delta\theta(0) =0$

$$\boxed{\theta(0,t) = 0}$$

#### Section 2: Solve for eigenvalues, eigenfunctions and plotting first three modes

Assuming the harmonic solution $\theta(x,t) = \phi(x) e^{i\omega t}$, we have the general solution 
\begin{equation*}
    \phi(x) = C_1 \sin(\beta x) + C_2 \cos(\beta x)
\end{equation*}
where $\beta = \omega \sqrt{\frac{I_{shaft}}{GJ}}$.
Apply boundary condition 
* At $x=0$, $C_2 = 0 \implies \phi(x) = C_1 \sin(\beta x)$.
* At $x=L$,
$$GJ \phi'(L) - I_{gear} \omega^2 \phi(L) = 0$$
$$GJ \beta C_1 \cos(\beta L) - I_{gear} \omega^2 C_1 \sin(\beta L) = 0$$
$$GJ \beta = I_{gear} \omega^2 \tan(\beta L)$$
Substitute $\omega^2 = \frac{\beta^2 GJ}{I_{shaft}}$:
$$GJ \beta = I_{gear} \left( \frac{\beta^2 GJ}{I_{shaft}} \right) \tan(\beta L)$$
$$1 = \frac{I_{gear}}{I_{shaft}} \beta \tan(\beta L)$$
Assume $J_{shaft} = I_{shaft} L$: 
$$L = \frac{I_{gear} L}{J_{shaft}} \beta \tan(\beta L)$$

so, the final transcedental characteristic equation is
$$\boxed{\beta L \tan(\beta L) = \frac{J_{shaft}}{I_{gear}}}$$
Let the mass ratio be $R = \frac{J_{shaft}}{I_{gear}}$ (Ratio of total shaft inertia to gear inertia). The equation is:$$x \tan(x) = R$$Where $x = \beta L$. Assume $R = 1.0$ means the shaft inertia is equals the gear inertia. 
$$\omega_n = (\beta_n L) \frac{1}{L} \sqrt{\frac{GJ}{I_{shaft}}}$$
And the normalized eigenfunctions are
$$\boxed{\phi_n(x) = \sin\left( (\beta_n L) \frac{x}{L} \right)}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# === 1. System Parameters ===
# R is the ratio of Total Shaft Inertia (J_shaft) to Gear Inertia (I_gear)
# R = 1.0 implies the shaft and the gear have the same moment of inertia.
R = 1.0 

# The transcendental equation we need to solve: x * tan(x) - R = 0
def characteristic_eq(x):
    return x * np.tan(x) - R

# === 2. Bisection Method ===
def find_root(func, x_min, x_max, tol=1e-6, max_iter=100):
    """
    Finds a root of func(x) = 0 in the interval [x_min, x_max]
    using the Bisection method.
    """
    if func(x_min) * func(x_max) > 0:
        print(f"Warning: Root might not exist in [{x_min}, {x_max}]")
        return None
    
    mid = (x_min + x_max) / 2.0
    for _ in range(max_iter):
        mid = (x_min + x_max) / 2.0
        if abs(func(mid)) < tol or (x_max - x_min) / 2.0 < tol:
            return mid
        
        if func(mid) * func(x_min) < 0:
            x_max = mid
        else:
            x_min = mid
            
    return mid

# === 3. Compute Eigenvalues ===
# We know the roots exist in specific intervals between asymptotes.
# Asymptotes are at pi/2, 3pi/2, 5pi/2...
# Roots are slightly before these asymptotes.

pi = np.pi
search_intervals = [
    (0.1, pi/2 - 0.1),      # Mode 1 Interval
    (pi, 3*pi/2 - 0.1),     # Mode 2 Interval
    (2*pi, 5*pi/2 - 0.1)    # Mode 3 Interval
]

eigenvalues = []
print(f"Solving for Mass Ratio R = {R}...")

for i, (low, high) in enumerate(search_intervals):
    root = find_root(characteristic_eq, low, high)
    if root:
        eigenvalues.append(root)
        print(f"Mode {i+1}: Beta*L = {root:.4f}")

# === 4. Plotting the Eigenfunctions ===
# Spatial domain x/L from 0 to 1
x_norm = np.linspace(0, 1, 500)

plt.figure(figsize=(10, 6), dpi=100)

styles = ['-', '--', '-.']
colors = ['blue', 'green', 'red']

for i, beta_L in enumerate(eigenvalues):
    # Calculate Mode Shape: phi(x) = sin(beta * x)
    # Using normalized coordinate: phi(x/L) = sin(beta*L * (x_norm))
    mode_shape = np.sin(beta_L * x_norm)
    
    # Normalize amplitude to 1 for clearer comparison
    mode_shape = mode_shape / np.max(np.abs(mode_shape))
    
    label_txt = f'Mode {i+1} ($\\beta L$={beta_L:.2f})'
    plt.plot(x_norm, mode_shape, linestyle=styles[i], color=colors[i], 
             linewidth=2, label=label_txt)

# Decoration
plt.title(f'Torsional Mode Shapes: Shaft with Tip Gear (Ratio $J_{{shaft}}/I_{{gear}}={R}$)')
plt.xlabel('Position along shaft ($x/L$)')
plt.ylabel('Normalized Twist Angle $\\phi(x)$')
plt.axhline(0, color='black', linewidth=1) # Zero line
plt.axvline(1, color='gray', linestyle=':', alpha=0.5) # Tip line

# Add annotation for the gear
plt.text(1.02, 0, 'Gear (Tip Mass)', rotation=90, verticalalignment='center')

plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Setting $x=L$ and dividing by $\beta^2$ and $\beta^3$:

Determinant of coefficient matrix

```python
import numpy as np

# 1. Define the Characteristic Equation Function
def cantilever_eq(x):
    """f(x) = 1 + cos(x)cosh(x) for the fixed-free beam."""
    return 1 + np.cos(x) * np.cosh(x)

# 2. Implement the Bisection Method
def bisection_solve(func, a, b, tol=1e-5, max_iter=500):
    """
    Finds a root of func within the interval [a, b] using the Bisection Method.
    Assumes func(a) and func(b) have opposite signs.
    """
    if func(a) * func(b) >= 0:
        print(f"Error: Bisection method requires f(a) and f(b) to have opposite signs. f({a})={func(a):.2f}, f({b})={func(b):.2f}")
        return np.nan

    c = a # Initialize c
    for i in range(max_iter):
        c = (a + b) / 2
        f_c = func(c)

        # Check for convergence
        if np.abs(f_c) < tol or (b - a) / 2 < tol:
            return c

        # Narrow the interval
        if func(a) * f_c < 0:
            b = c
        else:
            a = c

    # Return best estimate if max_iter reached
    return c

# 3. Define Initial Intervals (based on approximations (n - 0.5) * pi)
# We use slightly wider bounds to ensure sign change.
intervals = [
    (1.5, 2.0), # For the 1st root (~1.571)
    (4.5, 5.0), # For the 2nd root (~4.712)
    (7.5, 8.0)  # For the 3rd root (~7.854)
]

# 4. Find the Roots (beta_n L)
roots = []
for n, (a, b) in enumerate(intervals, 1):
    root = bisection_solve(cantilever_eq, a, b)
    roots.append(root)

beta_L_roots = np.array(roots)

# 5. Calculate Natural Frequency Factors
# w_n = (beta_n L)^2 * C, where C = sqrt(EI / (rho * A * L^4))
beta_L_squared = beta_L_roots**2

# 6. Print Results
print("="*60)
print("Results from Bisection Method (Tolerance 1e-5)")
print("="*60)
print(f"{'Mode (n)':<10} | {'Root (beta_n L)':<20} | {'(beta_n L)^2 Factor':<20} | {'Natural Frequency (w_n)':<30}")
print("-" * 60)
C_str = r"C"
for i in range(3):
    mode = i + 1
    root = beta_L_roots[i]
    factor = beta_L_squared[i]
    freq_expr = f"{factor:.4f} * C"
    print(f"{mode:<10} | {root:<20.5f} | {factor:<20.4f} | {freq_expr:<30}")
print("\nWhere C = sqrt(EI / (rho * A * L^4)).")
```

Zero Deflection: $W(0) = 0 \implies C_2 + C_4 = 0 \implies C_4 = -C_2$

Zero Slope: $W'(0) = 0 \implies C_1 + C_3 = 0 \implies C_3 = -C_1$

Applying Free End Boundary ConditionTo relate $C_1$ and $C_2$, we apply one of the free end boundary conditions at $x=L$. We use the condition that the bending moment is zero:

Zero Moment: $W''(L) = 0$

The second derivative of the reduced solution is:

$$W''(x) = C_1 \beta_n^2 (-\sin(\beta_n x) - \sinh(\beta_n x)) + C_2 \beta_n^2 (-\cos(\beta_n x) - \cosh(\beta_n x))$$
Setting $W''(L) = 0$ and dividing by $\beta_n^2$:

$$C_1 (-\sin(\beta_n L) - \sinh(\beta_n L)) + C_2 (-\cos(\beta_n L) - \cosh(\beta_n L)) = 0$$
Now, we solve this linear equation for the ratio of $C_1$ to $C_2$:

$$C_1 (\sin(\beta_n L) + \sinh(\beta_n L)) = -C_2 (\cos(\beta_n L) + \cosh(\beta_n L))$$
$$\implies C_1 = -C_2 \left[ \frac{\cos(\beta_n L) + \cosh(\beta_n L)}{\sin(\beta_n L) + \sinh(\beta_n L)} \right]$$

substitute this expression for $C_1$ back into the reduced mode shape equation

$$W_n(x) = \left[ -C_2 \frac{\cos(\beta_n L) + \cosh(\beta_n L)}{\sin(\beta_n L) + \sinh(\beta_n L)} \right] (\sin(\beta_n x) - \sinh(\beta_n x)) + C_2 (\cos(\beta_n x) - \cosh(\beta_n x))$$

Factor out $C_2$ and let $C_2 = C_{1n}$:

$$\boxed{W_n(x) = C_{1n} \left[ (\cos(\beta_n x) - \cosh(\beta_n x)) - \frac{\cos(\beta_n L) + \cosh(\beta_n L)}{\sin(\beta_n L) + \sinh(\beta_n L)} (\sin(\beta_n x) - \sinh(\beta_n x)) \right]}$$

### Part d:

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the first three roots (beta_n * L) of the characteristic equation
# These are the eigenvalues that define the mode shapes.
beta_L_roots = np.array([1.87510, 4.69409, 7.85476])

# 2. Define the normalized coordinate (xi = x/L) from 0 to 1
xi = np.linspace(0, 1, 100)

# 3. Mode Shape Plotting Function
def plot_mode_shapes(roots, xi):
    """Calculates and plots the first three mode shapes."""
    plt.figure(figsize=(10, 6))

    for n, beta_L in enumerate(roots, 1):
        # Calculate the ratio alpha_n required for the mode shape equation
        alpha_n = (np.cos(beta_L) - np.cosh(beta_L)) / (np.sin(beta_L) - np.sinh(beta_L))

        # Calculate the mode shape W_n(xi)
        # W_n(xi) = [ (cos(beta_L*xi) - cosh(beta_L*xi)) - alpha_n * (sin(beta_L*xi) - sinh(beta_L*xi)) ]
        W_n = (
            (np.cos(beta_L * xi) - np.cosh(beta_L * xi)) -
            alpha_n * (np.sin(beta_L * xi) - np.sinh(beta_L * xi))
        )

        # Normalize by the maximum absolute value for clean plotting (C_1n = 1)
        W_n_normalized = W_n / np.max(np.abs(W_n))

        # Plotting
        plt.plot(xi, W_n_normalized, label=rf'Mode {n} ($\beta L$ = {beta_L:.3f})', linewidth=2)


# Execute the plotting function
plot_mode_shapes(beta_L_roots, xi)


# Plot aesthetics
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
# plt.axvline(0, color='k', linestyle='-', linewidth=2) # Fixed end (xi=0)
plt.xlabel(r'Normalized Position, $\xi = x/L$', fontsize=12)
plt.ylabel(r'Normalized Deflection, $W_n(\xi)$', fontsize=12)
plt.title(r'First Three Mode Shapes of Fixed-Free Uniform Beam', fontsize=14)
# plt.legend(loc='lower right')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()
# # Save the figure
plt.savefig('fixed_free_mode_shapes.png')
```

and $\zeta = \frac{x}{L}$
2. find: 
$$m_{ij} = m_{ji} = \int_0^L m(x) \phi_i(x) \phi_j(x) dx$$

for $i,j = 1,2,...,n$

3. find:
$$k_{ij} = k_{ji} = \int_0^L EI(x) \frac{d^2\phi_i}{dx^2} \frac{d^2\phi_j}{dx^2} dx$$

for $i,j = 1,2,...$
4. establish $M\ddot{q} + Kq = 0$, and solve for corresponding eigenvalue problem $$\omega_n = (\beta_n L)^2 \sqrt{\frac{E I}{\rho A L^4}}$$, construct eigenfunction

Let $\zeta = \frac{x}{L}$ is the non-dimensional coordinate for $0<\zeta <1$, so $dx = L d\zeta$
For the three admissible function 
$$\phi_1(x) = \frac{x}{L},~ \phi_2(x) = \frac{x^2}{L^2},~ \phi_3(x) = \frac{x^3}{L^3}$$
$$\implies \phi_1(\zeta) = \zeta, ~\phi_2(\zeta) = \zeta^2, ~\phi_3(\zeta) = \zeta^3$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh

def solve_rayleigh_ritz():
    # 1. Define Matrices (Dimensional factors removed for eigenvalue calculation)
    # K_factor = EI / L^3
    # M_factor = rho * A * L
    
    K = np.array([
        [0.0, 0.0, 0.0],
        [0.0, 3.0, 4.0],
        [0.0, 4.0, 7.5]
    ])
    
    M = np.array([
        [5/24, 3/20, 7/60],
        [3/20, 7/60, 2/21],
        [7/60, 2/21, 9/112]
    ])
    
    # 2. Solve generalized eigenvalue problem: K * v = lambda * M * v
    # vals = omega^2 * (rho*A*L^4 / EI)
    evals, evecs = eigh(K, M)
    
    # Frequencies (omega_n)
    # omega = sqrt(val) * sqrt(EI / (rho*A*L^4))
    omegas_coeffs = np.sqrt(np.abs(evals))
    
    print("Dimensionless Frequency Coefficients (Beta_n L)^2:")
    for i, val in enumerate(omegas_coeffs):
        print(f"Mode {i}: {val:.4f}")

    # 3. Construct Mode Shapes
    # phi = [zeta, zeta^2, zeta^3]
    # W(zeta) = v[0]*zeta + v[1]*zeta^2 + v[2]*zeta^3
    
    zeta = np.linspace(0, 1, 100)
    phi1 = zeta
    phi2 = zeta**2
    phi3 = zeta**3
    
    plt.figure(figsize=(10, 6))
    
    modes_labels = ['Rigid Body (0 Hz)', '1st Flexible Mode', '2nd Flexible Mode']
    colors = ['gray', 'blue', 'red']
    styles = ['--', '-', '-']
    
    for i in range(3):
        # Extract eigenvector for mode i
        vec = evecs[:, i]
        
        # Calculate shape
        w = vec[0]*phi1 + vec[1]*phi2 + vec[2]*phi3
        
        # Normalize for plotting (max amplitude = 1)
        w_norm = w / np.max(np.abs(w))
        
        plt.plot(zeta, w_norm, label=f'{modes_labels[i]} ($\lambda$={omegas_coeffs[i]:.2f})', 
                 color=colors[i], linestyle=styles[i], linewidth=2)

    plt.title('Mode Shapes of Tapered Beam (Rayleigh-Ritz Method)')
    plt.xlabel(r'Normalized Position ($\zeta = x/L$)')
    plt.ylabel(r'Normalized Deflection $w(\zeta)$')
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1, linestyle='--')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    solve_rayleigh_ritz()
```
